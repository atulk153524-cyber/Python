### Q36. Import `students_scores.csv` as a pandas DataFrame `df`. Then convert the `math`, `english`, and `science` columns to separate NumPy arrays and print their shapes.

```python
# answer here
import pandas as pd

students_score = pd.read_csv('/content/students_scores.csv')
math_array = students_score['math'].to_numpy()
english_array = students_score['english'].to_numpy()
science_array = students_score['science'].to_numpy()

print(math_array.shape)
print(english_array.shape)
print(science_array.shape)
```

### Q37. Using NumPy on the arrays from Q36, compute the mean score of each subject (math, english, science) and print them.

```python
# answer here
print(math_array.mean())
print(english_array.mean())
print(science_array.mean())
```

### Q38. Using NumPy, compute each student’s **total score** (math + english + science) and print the `id` of the student with the highest total.

```python
# answer here
total_scores = math_array + english_array + science_array

top_student_index = np.argmax(total_scores)

print("Total scores:", total_scores)
print("ID of top student:", top_student_index)
```

### Q39. Using NumPy, standardize the `math` scores to **z-scores**: `(x - mean) / std` and print the first 5 standardized values.

```python
# answer here
mean = np.mean(math_array)
std = np.std(math_array)

# Standardize to z-scores
z_scores = (math_array - mean) / std

# Print first 5 standardized values
print(z_scores[:5])
```
