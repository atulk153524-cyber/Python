## 5. Merging and Concatenation

```python
# Example: Merge and Concat
df1 = pd.DataFrame({'ID': [1,2], 'Name': ['A', 'B']})
df2 = pd.DataFrame({'ID': [1,2], 'Score': [85, 90]})
merged = pd.merge(df1, df2, on='ID')
print(merged)

# Concat
df3 = pd.concat([df1, df2], axis=0, ignore_index=True)
print(df3)
```

## 6. Pivot Tables and GroupBy

```python
# Example: Pivot and GroupBy
df = pd.DataFrame({'Region': ['North', 'South', 'North'], 'Month': ['Jan', 'Jan', 'Feb'], 'Sales': [250, 300, 400]})

pivot = df.pivot_table(values='Sales', index='Region', columns='Month', aggfunc='sum', fill_value=0)
grouped = df.groupby('Region')['Sales'].sum()
print(pivot)
print()
print(grouped)
```

### Practice Questions
1. Group the DataFrame by 'Month' and find total sales.
2. Create a pivot table to show Region-wise total sales.
3. Group data by 'Region' and count number of entries.

```python
# Answer here
data3 = {
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Jan', 'Feb', 'Mar'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'West', 'East', 'North'],
    'Sales': [2000, 1500, 2200, 1800, 2100, 2300, 2500, 1900, 2700]
}

data3_dataframe = pd.DataFrame(data3)

print()
print("-------------------Month wise total Sales-----------------")
print()

monthly_sales = data3_dataframe.groupby('Month')['Sales'].sum()
print(monthly_sales)

print()
print("-----------------Region wise total sales-------------------")
print()

pivot_sales = data3_dataframe.pivot_table(values='Sales', index='Region', aggfunc='sum')
print(pivot_sales)

print()
print("-----------------Number of entries in all region")
print()

region_count = data3_dataframe.groupby('Region').size()
print(region_count)
```

1. Import a dataset and clean all missing values, rename columns properly, and export it back to Excel.

Dataset- https://drive.google.com/file/d/1kw_Sdib3WCo_acQYA851poCUdKcWgfYs/view?usp=sharing

```python
# Answer here
messy = pd.read_csv('/content/messy_data.csv')

print("--------------data before cleaning-------------")
print()
print()

print(messy)

# filling the empty values
for i in messy.columns:
  if pd.api.types.is_numeric_dtype(messy[i]):
    messy[i] = messy[i].fillna(messy[i].mean())

  else:
    messy[i] = messy[i].fillna("Unknow")

# changing the column name
messy = messy.rename(columns={
    'Age(Years)': 'Age',
    'Sales( Rs)': 'Sales'
})

print()
print("--------------data after cleaning-------------")
print()
print(messy)

messy.to_excel('messy.xlsx', index=False)
```

2. Combine multiple CSV files from a folder into one DataFrame and find top 5 entries by a numeric column like 'Sales'.

Dataset- https://drive.google.com/drive/folders/1zgJs_6r8Aa_-93YXmYXfxOK5L4hVUI9Q?usp=sharing

```python
# Answer here
jan = pd.read_csv('/content/sales_jan.csv')
feb = pd.read_csv('/content/sales_feb.csv')
mar = pd.read_csv('/content/sales_mar.csv')

print()
print("------------------All Sales------------------")
print()

all_sales = pd.concat([jan, feb, mar], axis=0, ignore_index=True)
print(all_sales)

print()
print("------------------Top 5 Sales-----------------")
print()

top5_sales = all_sales.sort_values(by='Sales', ascending=False).head(5)
print(top5_sales)
```
