### Tuples Q20

**Problem:** Check the memory size of a tuple versus a list using `sys.getsizeof()`.

**Expected answer should look like (example):**
```
e.g., tuple_size < list_size (often)
```

````python
# write your code here
import sys

my_list = [1, 2, 3, 4, 5]
my_tuple = (1, 2, 3, 4, 5)

# Get memory size
list_size = sys.getsizeof(my_list)
tuple_size = sys.getsizeof(my_tuple)

print("List size:", list_size)
print("Tuple size:", tuple_size)

# Compare
print("Tuple is smaller than list:", tuple_size < list_size)

#  Explanation

# sys.getsizeof(obj) → returns the memory size in bytes of the object.

# Tuples are generally smaller than lists because they are immutable and require less overhead.

# Lists have extra memory for dynamic resizing and mutability.
````
